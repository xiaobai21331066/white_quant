# 数据分析

```python
import pickle
import pathlib
import os
from collections import defaultdict
```

```python
strategy_name = 'golden_cross'
result_dir = pathlib.Path('../white_quant/result/{}'.format(strategy_name))
```

```python
data_result = defaultdict(None)
for item in result_dir.iterdir():
    print(item.name.split('.')[0])
    with item.open('rb') as f:
        data_result[item.name.split('.')[0]] = pickle.load(f)
```

## summary数据分析

```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
summary_index = []
summary_data= []
for key, value in data_result.items():
    index = summary_index.append(key)
    summary_data.append(value['sys_analyser']['summary'])
summary_df = pd.DataFrame(summary_data, index=summary_index)
summary_df
```
